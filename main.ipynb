<h2>Data Cleaning</h2>

<h4>Application</h4>

```python
#drop a row iff all the columns are empty
app_df.drop_duplicates(inplace=True)

#rows will be dropped having na value in Rating Column(other columns(na or non-na) doesn't matter)
app_df=app_df.dropna(subset=['Rating'])

#rows will be dropped having na value in App Column
app_df.drop_duplicates(subset='App',inplace=True)

#filling na values with mode value of the respective column
for column in app_df.columns:
    app_df[column].fillna(app_df[column].mode()[0],inplace=True)

#rating more than 5 are removed
app_df=app_df[app_df['Rating']<=5]
```

<h4>Review</h4>

```python
#Delete rows having na value in Translated_column
rev_df.dropna(subset='Translated_Review', inplace=True)
```

<h2>DataSet Transformation</h2>

```python
app_df.head()
```

```python
rev_df.head()
```

```python
app_df.info()
```

```python
#simple type conversion
app_df['Installs']=app_df['Installs'].str.replace('+','').str.replace(',','').astype(int)
app_df['Price']=app_df['Price'].str.replace('$','').astype(float)

#M and k removed and type conversion of Size
for i in app_df.index:
    app_df.at[i,'size']=app_df.at[i,'Size'][-1]
app_df['Size']=app_df['Size'].str.replace('M','').str.replace('k','').str.replace('Varies with device','0').astype(float)

#instead of k and M converting in actual number
for i in app_df.index:
    if(app_df.at[i,'size']=='k'):
        app_df.at[i,'Size']=app_df.at[i,'Size']*(2**10)
    elif (app_df.at[i,'size']=='M'):
        app_df.at[i,'Size']=app_df.at[i,'Size']*(2**20)
app_df.drop(columns=['size'],inplace=True)
app_df
```

```python
#Categorization of Rating
def review_rating(rating):
    if rating >=4:
        return 'top rated app'
    elif rating >= 3:
        return 'above average'
    elif rating >=2:
        return 'Average'
    else:
        return 'below average'
app_df['Rating_Group']=app_df['Rating'].apply(review_rating)

#Calculating revenue for each app
app_df['Revenue']=app_df['Installs']*app_df['Price'].astype(float)
app_df.head(3)
```

<h3>Both Data Merged</h3>

```python
df=pd.merge(app_df,rev_df,on='App', how='inner')
df.head(2)
```
